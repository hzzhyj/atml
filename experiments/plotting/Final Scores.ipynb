```python
import numpy as np
import os
import sys
import scipy.stats
import pandas as pd

module_path = os.path.abspath(os.path.join('../..'))
sys.path.append(module_path+"/models")
sys.path.append(module_path+"/train")
sys.path.append(module_path+"/datasets")
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return f"{m*100:04.2f}% +- {h*100:01.1f}%"
```

```python
folders = ["normal_dataset_bernoulli_fc_models", 
           'uniform_noise_dataset_bernoulli_fc_models',
           'generated_noise_bernoulli_fc_models']
```

```python
scores = []
for folder in folders:
    model_names = np.load(folder+"_model_names.npy")
    betavae_metric_accuracies = np.load(folder+"_betavaemetric_scores.npy")
    factorvae_metric_accuracies = np.load(folder+"_factorvaemetric_scores.npy")
    mig_scores = np.load(folder+"_mig_scores.npy")
    for i, model in enumerate(model_names):
        sc = {}
        sc['folder'] = folder
        sc['model'] = model 
        try:
            sc['betavae'] = mean_confidence_interval(betavae_metric_accuracies[i])
            sc['factorvae'] = mean_confidence_interval(factorvae_metric_accuracies[i])
            sc['mig'] = mean_confidence_interval(mig_scores[i])
        except:
            pass
        scores.append(sc)
```

```python
df = pd.DataFrame(scores).sort_values(by=['folder', 'model'], axis=0)
```
